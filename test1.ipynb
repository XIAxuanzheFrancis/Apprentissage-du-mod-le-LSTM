Les paramètres de détection sont collectés pour chaque capteur et un fichier est généré avec la date, les paramètres de détection, et le paramètre de connexion pour la fusion des deux fichiers est l'identification du capteur.

```python
import matplotlib.pyplot as plt
import os
```

```python

merged_df = pd.merge(data, infoCap, on='id_identification', how='left')

output_dir = 'sensor_plots'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

groups = merged_df.groupby('id_identification')
for sensor_id, group_data in groups:
    # Extract unique sensors for this sensor id
    sensors = group_data['Sensor'].unique()
    # Create a new DataFrame with columns Date and all sensors for this sensor id
    sensor_df = pd.DataFrame(columns=['Date'] + list(sensors))
    # Iterate through each date
    for date, date_group in group_data.groupby('Date'):
        # Create a dictionary to store values for each sensor at this date
        sensor_values = {'Date': date}
        # Iterate through each sensor
        for sensor in sensors:
            # Check if there are any values for this sensor at this date
            if not date_group[date_group['Sensor'] == sensor].empty:
                # Extract the value for this sensor at this date
                value = date_group[date_group['Sensor'] == sensor]['Valeurs'].iloc[0]
                # Store the value in the dictionary
                sensor_values[sensor] = value
        # Append the sensor values for this date to the DataFrame
        sensor_df = sensor_df.append(sensor_values, ignore_index=True)
    # file_name = f"{output_dir}/{sensor_id}.png"
    # plt.savefig(file_name)
    # plt.close()
    # print(f"{file_name} generated")
    # Generate the filename, formatted as id_identification.csv
    file_name = f"{sensor_id}.csv"
    # Save the DataFrame to a CSV file
    sensor_df.to_csv(file_name, index=False)
    print(f" {file_name} generated")
    # Convert Date columns to datetime type
    sensor_df['Date'] = pd.to_datetime(sensor_df['Date'])
    sensor_df = sensor_df.set_index('Date')
    plt.figure(figsize=(10, 6))
    for sensor in sensors:
        plt.plot(sensor_df.index, sensor_df[sensor], label=sensor)

    plt.title(f'Sensor Data for id_identification {sensor_id}')
    plt.xlabel('Date')
    plt.ylabel('Values')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
values = data.values
dates = data.index
groups = [0,1,2,3,4,5,6,7]
i = 1

plt.figure(figsize=(20,30))
for group in groups:
    plt.subplot(len(groups), 1, i)
    plt.plot(dates,values[:, group])
    plt.title(data.columns[group], y=0.5, loc='right')
    i += 1
plt.show()
```
